```python
import pandas as pd 
import jsonlines 

def prodigy_result_to_pandas(path):
    
    with jsonlines.open(path) as reader:

        response_list = []

        for response in reader:

            response_dict = {}

            response_dict['idannonce'] = response['id']
            response_dict['annotator'] = response['_session_id']

            if 'selected_ids' in response.keys():
                response_dict['answer'] = response['selected_ids']
            else:
                response_dict['answer'] = []

            response_dict['status'] = response['answer']
            response_dict['proposal_id_list'] =  [proposal['id'] for proposal in response['options']]

            response_list.append(response_dict)   
            
    return(pd.DataFrame(response_list))
```

```python
df = prodigy_result_to_pandas('../data/srd_annotated_1.jsonl')
```
